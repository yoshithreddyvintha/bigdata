```
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer

# Load datasets
red_wine_data = pd.read_csv('/content/winequality-red.csv', sep=';')
white_wine_data = pd.read_csv('/content/winequality-white.csv', sep=';')

# Add column to distinguish wine type
red_wine_data['wine_type'] = 'red'
white_wine_data['wine_type'] = 'white'

# Combine datasets
combined_data = pd.concat([red_wine_data, white_wine_data], ignore_index=True)

# Display basic dataset information
print(combined_data.info())
print(combined_data.describe())

# Check for missing values
print("Missing Values:")
print(combined_data.isnull().sum())

# Analyze distribution of quality ratings
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8, 5))
sns.countplot(x='quality', data=combined_data, palette="viridis")
plt.title('Distribution of Wine Quality Ratings')
plt.xlabel('Quality')
plt.ylabel('Count')
plt.show()

# Correlation heatmap
plt.figure(figsize=(12, 8))
correlation_matrix = combined_data.corr(numeric_only=True)
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()

# Preprocessing: Separate features and target
X = combined_data.drop(columns=['quality'])
y = combined_data['quality']

# One-hot encode 'wine_type' and standardize numeric features
preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), X.select_dtypes(include=['float64']).columns),
        ('cat', OneHotEncoder(drop='first'), ['wine_type'])
    ]
)

X_processed = preprocessor.fit_transform(X)

# Split the dataset
X_train, X_test, y_train, y_test = train_test_split(X_processed, y, test_size=0.2, random_state=42, stratify=y)

# Check dataset shapes
print(f"Training set shape: {X_train.shape}, {y_train.shape}")
print(f"Testing set shape: {X_test.shape}, {y_test.shape}")
```

```
from sklearn.utils import resample

# Create a DataFrame for easy manipulation
train_data = pd.DataFrame(X_train)
train_data['target'] = y_train

# Resample each class manually
oversampled_data = []
for cls in train_data['target'].unique():
    class_data = train_data[train_data['target'] == cls]
    if len(class_data) < min_samples:  # Oversample if less than min_samples
        class_data = resample(class_data, replace=True, n_samples=min_samples, random_state=42)
    oversampled_data.append(class_data)

# Combine oversampled data
balanced_data = pd.concat(oversampled_data)

# Separate features and target again
X_train_resampled = balanced_data.drop(columns=['target']).to_numpy()
y_train_resampled = balanced_data['target'].to_numpy()
```
